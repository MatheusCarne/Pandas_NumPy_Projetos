# d) quantos acidentes com feridos graves aconteceram na Paraíba em 2019?

```python
df2= pd.DataFrame(df19[['uf', 'feridos_graves']])
df_pb = df2[df2.uf.isin(['PB'])]
df_feridos=df_pb['feridos_graves'] > 0
qtd_feridos_pb=df_pb[df_feridos]
qtd_acidentes = qtd_feridos_pb['uf'].value_counts()
qtd_acidentes
```

# Tarefa 3 Construindo novos DataFrames

### Primeiro com a função concat fazemos a união dos 3 DataFrames df18, df19 e df20.

```python
df_dados=pd.concat([df18,df19,df20],axis=0,	ignore_index=True)
df_dados
```

# a) ranking de acidentes por estado para todos os anos combinados

```python
ranking=df_dados['uf'].value_counts()
ranking.rank
df3=pd.DataFrame(ranking)
df3.rename(columns={'uf':'acidentes'}, inplace=True)
df3.index.names = ['Municipios']
df3.to_csv('Ranking.csv', sep='\t', encoding='utf-8')
df3
```

# b) acidentes por dia da semana para todos os anos combinados

```python
acid=df_dados['dia_semana'].value_counts()
acidentes=pd.DataFrame(acid)
acidentes.rename(columns={'dia_semana':'acidentes'}, inplace=True)
acidentes.index.names = ['Dias da Semana']
acidentes.to_csv('Dias_da_Semana.csv', sep='\t', encoding='utf-8')
acidentes
```

# Tarefa 4 Ocorrencia de acidentes

# a) qual a causa mais frequente e a mais rara de acidentes registrados nos três anos?

```python
df_completo = pd.concat([df20,df19,df18])
causas_acidentes = df_completo['causa_acidente'].value_counts()
causas_acidentes = pd.DataFrame(causas_acidentes,columns=["causa_acidente"])
causas_acidentes = causas_acidentes.rename(columns={'causa_acidente': 'Quantidade de Acidentes'})
causas_acidentes['causa_acidente'] = causas_acidentes.index
causas_acidentes = causas_acidentes.set_index(['causa_acidente'])
causas_acidentes.head(1)
```

### A causa mais frequente é falta de atenção á condução com 73231 casos.

### E as causas mais raras com somente 1 ocorrencia esta na tabela abaixo.
